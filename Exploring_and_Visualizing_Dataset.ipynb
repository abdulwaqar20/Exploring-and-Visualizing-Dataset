#Data Visualization

Scatter Plot

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(
    data=iris,
    x="sepal_length",
    y="sepal_width",
    hue="species"
)
plt.title("Sepal Length vs Sepal Width")
plt.show()
```

## Scatter Plot: Sepal Length vs Sepal Width

This scatter plot shows the relationship between sepal length and sepal width for
different Iris species.

Each point represents a flower sample, and colors indicate the species.
Setosa forms a clearly separated cluster, while versicolor and virginica show
overlap. This indicates that sepal features alone are not sufficient to fully
separate all species.

Histograms

```python
iris.hist(figsize=(10, 8))
plt.suptitle("Feature Distributions", fontsize=16)
plt.show()
```

## Histograms: Feature Distributions

The histograms display the distribution of values for each numerical feature in
the dataset.

Sepal length and sepal width show relatively normal distributions with overlap.
Petal length and petal width show clearer separation and multiple peaks, indicating
significant variation among species.

Box Plots(Outlier Detection)

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=iris)
plt.title("Box Plot of Iris Features")
plt.show()
```

## Box Plot: Iris Feature Comparison

This box plot summarizes the distribution of all numerical features, showing
medians, interquartile ranges, and potential outliers.

Petal features exhibit larger variability compared to sepal features.
A few outliers are visible in sepal width, and petal measurements show clearer
distinction in value ranges.

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x="species", y="petal_length", data=iris)
plt.title("Petal Length by Species")
plt.show()
```

## Box Plot: Petal Length by Species

This box plot compares petal length across different Iris species.

Setosa has the smallest petal lengths with minimal variation.
Versicolor has medium petal lengths, while virginica has the largest.
There is very little overlap between species, making petal length a strong
discriminative feature.
